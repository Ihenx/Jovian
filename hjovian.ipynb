```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    for idx, value in col_series[col_series.notnull()].iteritems():
        for option in value.split(';'):
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
                
            result_df.at[idx, option] = True
            
    return result_df[options]
```

```python
dev_type_df = split_multicolumn(survey_df.DevType)
dev_type_df
```

```python
dev_type_df_totals = dev_type_df.sum().sort_values(ascending=False)
dev_type_df_totals
```

```python
sum_of_dev =dev_type_df_totals.sum() 
sum_of_dev
```

```python
dev_pop = dev_type_df.mean().sort_values(ascending = False)*100
dev_pop
```

## what are the most popular programming languages in 2020

```python
language_worked_with_df = split_multicolumn(survey_df.LanguageWorkedWith)
language_worked_with_df
```

```python
languages_worked_percentages= language_worked_with_df.mean().sort_values(ascending=False)*100
languages_worked_percentages
```

```python
plt.figure(figsize=(20,10))
plt.title('Language used in the past year')
plt.xlabel("Count")
sns.barplot(x=languages_worked_percentages, y=languages_worked_percentages.index, palette="crest");
```

```python
language_desired = split_multicolumn(survey_df.LanguageDesireNextYear)
language_desired
```

```python
language_desired_percentages = language_desired.mean().sort_values(ascending=False)* 100
language_desired_percentages
```

```python
plt.figure(figsize=(20,10))
plt.title("language desired next year")
sns.barplot(x= language_desired_percentages, y = language_desired_percentages.index, palette='inferno');
```

### Which is the most loved languages ie a high percentage of people who have  used the language and want want 


```python
language_loved_df = language_worked_with_df & language_desired
language_loved_df
```

```python
languages_loved_percentages = (language_loved_df.sum()*100 / language_worked_with_df.sum()).sort_values(ascending=False)
languages_loved_percentages
```

```python
plt.figure(figsize=(12,12))
sns.barplot(x =languages_loved_percentages, y = languages_loved_percentages.index, palette='gist_heat')
plt.title("Most loved language")
plt.xlabel("count");
```
